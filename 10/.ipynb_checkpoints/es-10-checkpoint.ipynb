### Exercise 10.1
We improve the SA algorithm by adding parallelization: using MPI, many walkers search for the solution independently. Periodically, the first node gathers information about the results of all other nodes, in order to keep track of the length of the overall best path found among all parallel searches. The graph tracks the evolution of this overall best, since there is no single "istantaneous" path.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as lines


path = np.loadtxt("main/res/box100/1/final-path.dat")

x2s = path[:,0]; y2s = path[:,1]

fig2, ax2 = plt.subplots()
ax2.scatter(x2s, y2s)
l2 = lines.Line2D(x2s,y2s)
ax2.add_line(line=l2)
plt.title("Final Path")

fig3, ax3 = plt.subplots()

his2 = np.loadtxt("main/res/box100/1/len-his.dat.1")
hx2 = his2[:,0]; hy2 = his2[:,1]
ax3.plot(hx2,hy2, label='non-parallel')

his = np.loadtxt("main/res/box100/1/len-his.dat.20")
hx = his[:,0]; hy = his[:,1]
ax3.plot(hx,hy, label='parallel (20 nodes)')
plt.title("length of best path")
plt.xlabel("iteration step")
plt.ylabel("length")


ax3.legend()
plt.show()
```

This version of the code does not show significant improvement in the speed of the algorithm, even with more complex problems (above, 100 cities instead of 30), or with a lot of parallel nodes. After all, with the parameters used, the convergence of the algorithm seems to be more strongly tied to the temperature descent speed than to the random factor of the search. So, in a way, the parallel processes all have to wait for their temperature to decrease before yielding a good result.
