### Exercise 09.1
We use a gentic algorithm to find solutions of the Traveling Salesman problem (TSP). Our algorithm utilizes a batch of $M = 900$ paths, initially chosen at random as permutations of the $N = 30$ TSP cities. 

The algorithm then uses mutation functions and crossovers to look for better solutions, starting from the ones in the batch. At each step of the algorithm, the top 5% of the batch population is kept untouched, while the top 50% of the remainder is substituted by mutations of themselves, and the bottom 50% is eliminated completely and substituted by either mutations of the top elements or crossovers between top and middle elements.
The mutation functions are:
- random swap of two cities
- shift of a contiguous block of cities
- shift of all cities (helpful for ergodicity)
- path inversion
- full reshuffling

The crossover function works by cutting the first path in half, and completing the first half with the remaining cities in the order in which they appear in the second path.

The purpose of this setup is to try to avoid destructive mutations but still leave room for the top elements to evolve, using the slots freed by the bottom ones, in order to avoid getting stuck in local minima.

We tested the algorithm on a circular configuration of cities as well as on a uniform random one. The graphs show the final path calculated by the algorithm, and the length of the best path as a function of the iteration of the algorithm.

### Circle city

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as lines

x2s, y2s = np.loadtxt("main/res/final/circle/final-path.dat.yes", unpack='true')

fig2, ax2 = plt.subplots()
ax2.scatter(x2s, y2s)
l2 = lines.Line2D(x2s,y2s)
ax2.add_line(line=l2)

fig, ax = plt.subplots()
toplx,toply = np.loadtxt("main/res/final/circle/len-his.dat", unpack='true')
ax.plot(toplx,toply)
ax.set_xlabel('generation')
ax.set_ylabel('length')

plt.show()
```
