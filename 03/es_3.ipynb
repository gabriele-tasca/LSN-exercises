### Exercise 03.1
We use Monte-Carlo simulation for a geometric Brownian motion (GBM) model for stock prices to evaluate the prices of call and put European options.

The stochastic equation of motion for the GBM is known analytically, and it allows to extract the probability distribution for the price at any time $t$ knowing the price at any other time $t_0$. For both call and put options, we sampled the price evolution over the time interval covered by the option both in a single computation from $t_0$ to $t$, and with a succession of 100 discrete jumps covering 1/100 of the interval.

The resulting prices are then compared with the analytical solution for the option price itself, as calculated by Black and Scholes.

We use the following parameters:
By using the following parameters: 
- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$

With these parameters, the Black-Scholes solution yields the values:

```python
import scipy
from scipy.special import erf
from scipy import log, exp, sqrt

# data
S0 = 100.
K = 100.
T = 1.
r = 0.1
sigma = 0.25
t=0

def N(x):
    return 0.5 * (1. + erf(x / sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * sqrt(T)) * (log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * sqrt(T)
    C = S0 * N(d1) - K * exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * exp(-r * T) * (N(d2)-1.)
    return C, P

if __name__ == "__main__":
    call, put = black_scholes(S0, K, T, r, sigma)
    print(" == BLACK-SCHOLES ==")
    print("call: ", call)
    print("put: ", put)
```

The results of the Monte-Carlo calculations are presented below, for the four cases:
- Call option, direct sampling:

```python
import numpy as np
import matplotlib.pyplot as plt
import math
with open('1/call.out') as resultsfile: 
    N = int(resultsfile.readline().split()[0])
    print("Number of blocks: N =", N)
    resultsfile.readline() #kill the line with the comments
    resultsfile.readline() #kill the first line, where the mean stdev can't be computed
    extra_killed_lines = 100
    realN = N-2-extra_killed_lines
    for i in range(0,extra_killed_lines):
        resultsfile.readline()
    numbers = np.zeros(realN)
    means_v = np.zeros(realN)
    errors_v = np.zeros(realN)
    for i in range(0,realN):
        temp = resultsfile.readline().split()
        numbers[i] = i
        means_v[i] = temp[0]
        errors_v[i] = temp[1]

    plt.errorbar(numbers, means_v, yerr=errors_v, errorevery=100)
    plt.xlabel('Total random throws')
    plt.ylabel('observed value of I')
    plt.grid(True)
    plt.axhline(call,label="Black-Scholes solution", color="red")
    plt.legend()
    plt.show()
    
    print("Final estimation:",means_v[realN-1],"+-",errors_v[realN-1])
```
