### Exercise 05.1
We use a Monte Carlo algorithm with Metropolis sampling to study the 1s and 2p wavefunctions of the hydrogen atom.
For each wavefunction, we plot the sampling of the 3D probability function, and the data-blocking estimation of the mean value of $r$, for which we know the analytical values (in units of $a_0$):

$\left\langle r \right\rangle_{\Psi_{1,0,0}} = \dfrac{3}{2}$ 

$\left\langle r \right\rangle_{\Psi_{2,1,0}} = 5$

#### 1s wavefunction:

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
nmax = 2000
data = np.loadtxt("metro1s/smpl.out", skiprows=1)
X = data[:nmax,0]
Y = data[:nmax,1]
Z = data[:nmax,2]
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.show()
```

```python
a, b = np.loadtxt("metro1s/h.out",unpack=True,skiprows=1)
n = np.arange(1,len(a)+1)
plt.errorbar(n, a, yerr=b) ; plt.axhline(1.5, c="red")
plt.show()
print("last result.", a[-1], "+-", b[-1])
```

#### 2p wavefunction:

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
nmax = 5000
data = np.loadtxt("metro2p/smpl.out", skiprows=1)
X = data[:nmax,0]
Y = data[:nmax,1]
Z = data[:nmax,2]
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.show()
```

```python
a, b = np.loadtxt("metro2p/h.out",unpack=True,skiprows=1)
n = np.arange(1,len(a)+1)
plt.errorbar(n, a, yerr=b) ; plt.axhline(5, c="red")
plt.show()
print("last result.", a[-1], "+-", b[-1])
```

The mean value results are reasonably compatible with the known analytical values, and the distribution of the sampled points resemble the familiar shapes of the spherical harmonics.

We did not equilibrate the algorithm before performing our calculations. In the case of a simple 1D probability distribution each metropolis step rewrites completely the "state" of the system. Since the Metropolis algorithm is a Markov chain, this means that any previous equilibration time would have no effect, except for choosing a "good" starting point. In this case $ x = 0 $ is a reasonable and safe choice, so there is no need for equilibration at all.
