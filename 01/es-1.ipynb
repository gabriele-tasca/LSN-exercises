### Exercise 01.1
We test the accuracy of our random number generator. For each block of $L$ random throws we compute the mean value, and repeat the procedure for $N$ blocks to extract the standard deviation of the mean. The mean value is compatible with the expected value of $\frac{1}{2}$.

```python
import numpy as np
import matplotlib.pyplot as plt

with open('1.1/results.out') as resultsfile: 
    N = int(resultsfile.readline()) #number of blocks
    throws_in_block = int(resultsfile.readline()) 
    total_throws = int(resultsfile.readline()) #equal to throws_in_block * N
    print("Number of blocks: N = ", N)
    print("Random throws per block: L = ", throws_in_block)
    print("Total random throws: M = ", total_throws)
    resultsfile.readline() #kill the first line, where the mean stdev can't be computed
    numbers = np.zeros(N-1)
    means_v = np.zeros(N-1)
    errors_v = np.zeros(N-1)
    for i in range(0,N-1):
        temp = resultsfile.readline().split()
        numbers[i] = temp[0]
        means_v[i] = temp[1]
        errors_v[i] = temp[2]

    plt.errorbar(numbers*throws_in_block, means_v-0.5, yerr=errors_v, errorevery=10)
    plt.xlabel('Total random throws')
    plt.ylabel('<r>-1/2')
    plt.grid(True)
    plt.show()
```

With the same method used in the previous point, we check that the variance is compatible with the expected $\frac{1}{12}$. The variance is calculated as the square of the mean of the deviations from the (exactly known) mean value.

```python
with open('1.2/results.out') as resultsfile: 
    N = int(resultsfile.readline()) #number of blocks
    throws_in_block = int(resultsfile.readline()) 
    total_throws = int(resultsfile.readline()) #equal to throws_in_block * N
    print("Number of blocks: N = ", N)
    print("Random throws per block: L = ", throws_in_block)
    print("Total random throws: M = ", total_throws)
    resultsfile.readline() #kill the first line, where the mean stdev can't be computed
    numbers = np.zeros(N-1)
    means_v = np.zeros(N-1)
    errors_v = np.zeros(N-1)
    for i in range(0,N-1):
        temp = resultsfile.readline().split()
        numbers[i] = temp[0]
        means_v[i] = temp[1]
        errors_v[i] = temp[2]

    plt.errorbar(numbers*throws_in_block, means_v-(1./12.), yerr=errors_v, errorevery=100)
    plt.xlabel('Total random throws')
    plt.ylabel('sigma^2  - 1/12')
    plt.grid(True)
    plt.show()
```

We computed the chi-squared $\chi^2$ test N times, dividing the $[0,1]$ interval in $M$ bins, and using $L$ random numbers each time. With the same method as in the above points, we verified that the $\chi^2$ settles on a value compatible with a value of $\chi^2 = M$.

```python
with open('1.3/results.out') as resultsfile: 
    N = int(resultsfile.readline().split()[0]) #number of blocks
    bins = int(resultsfile.readline().split()[0]) 
    throws_per_bin = int(resultsfile.readline().split()[0]) #equal to throws_in_block * N
    print("Number of blocks: N = ", N)
    print("Number of bins: M = ", bins)
    print("Throws per block: L = ", throws_per_bin)
    resultsfile.readline() #kill the line with the comments
    resultsfile.readline() #kill the first line, where the mean stdev can't be computed
    numbers = np.zeros(N-1)
    chis = np.zeros(N-1)
    means_v = np.zeros(N-1)
    errors_v = np.zeros(N-1)
    for i in range(0,N-1):
        temp = resultsfile.readline().split()
        numbers[i] = temp[0]
        chis[i] = temp[1]
        means_v[i] = temp[2]
        errors_v[i] = temp[3]

    plt.plot(numbers*throws_per_bin, chis)
    plt.xlabel('Total random throws')
    plt.ylabel('chi squared')
    plt.grid(True)
    plt.show()
        
    plt.errorbar(numbers*throws_per_bin, means_v-(100), yerr=errors_v)
    plt.xlabel('Total random throws')
    plt.ylabel('progressive mean of chi squared - M')
    plt.grid(True)
    plt.show()
```

### Exercise 01.2
We test the validity of the Central Limit Theorem for a sum of random variables $S_N = \frac{1}{N}\sum_{i=1}^N x_i$, where $x_i$ is a uniform, exponential, or Lorentzian variable, and $N = 1, 2, 10, 100$. Each histogram represents the distribution of $M = 10^4$ istances of $S_N$, and the four colored curves correspond to the four different values of $N$.

```python
numbers = ["1","2","10","100"]
name = "unif"
vector = [None] * len(numbers)
for j in range(0,len(numbers)):
    with open("2/"+name+"_"+numbers[j]+".out") as currentfile:
        blocks = int(currentfile.readline().split()[0])
        throws = int(currentfile.readline().split()[0])
        vector[j] = np.zeros(blocks)
        for kappa in range(0,blocks):
            vector[j][kappa] = currentfile.readline()

for i in range(0, len(numbers)):
    plt.hist(vector[i],60, alpha=0.7)
plt.ylabel('Occurrences')
plt.show()   
```

```python
name = "exp"
vector = [None] * len(numbers)
for j in range(0,len(numbers)):
    with open("2/"+name+"_"+numbers[j]+".out") as currentfile:
        blocks = int(currentfile.readline().split()[0])
        throws = int(currentfile.readline().split()[0])
        vector[j] = np.zeros(blocks)
        for kappa in range(0,blocks):
            vector[j][kappa] = currentfile.readline()

for i in range(0, len(numbers)):
    plt.hist(vector[i],100, alpha=0.7)
plt.ylabel('Occurrences')
plt.show()   
```

```python
import numpy as np
abs_value_treshold = 20;

name = "lorentz"
vector = [None] * len(numbers)
for j in range(0,len(numbers)):
    with open("2/"+name+"_"+numbers[j]+".out") as currentfile:
        blocks = int(currentfile.readline().split()[0])
        throws = int(currentfile.readline().split()[0])
        vector[j] = np.zeros(blocks)
        for kappa in range(0,blocks):
            temp = float(currentfile.readline())
            if abs(temp) < abs_value_treshold:
                vector[j][kappa] = temp 
            else:
                vector[j][kappa] = 0.
                
for j in range(0,len(vector)):
    vector[j] = vector[j][ vector[j] != 0 ]  # splicing out all 0's. we trust that the generator 
    # never yields an exact zero.
    
print("absolute value treshold: ",abs_value_treshold)
for i in range(0, len(numbers)):
    plt.hist(vector[i],100, alpha=0.7)
plt.ylabel('Occurrences')
plt.show()   
```

We note that in the Lorentzian case, the histogram representation is very uninformative unless we filter out numbers with an absolute value over a certain treshold. Even if most values are centered around zero, it is still relatively easy to occasionally find very large numbers, and the histogram is forced to stack all of the "normal" values together in a single bin. This behavior is related to the infinite variance of the Lorentz distribution.

More interestingly, we note that the Lorentz distribution is stable: the shape of $S_N$ remains the same as $N = 1$, i. e. simply Lorentzian, for any value of $N$. 

### Exercise 01.3
We simulate the Buffet experiment for the evaluation of $\pi$. See the c++ code for implementation details.

```python
import math
with open('3/results.out.1') as resultsfile: 
    N = int(resultsfile.readline().split()[0]) #number of blocks
    throws = int(resultsfile.readline().split()[0])
    print("Number of blocks: N =", N)
    print("Number of throws in each block: L =", throws)
    resultsfile.readline() #kill the line with the comments
    resultsfile.readline() #kill the first line, where the mean stdev can't be computed
    numbers = np.zeros(N-1)
    means_v = np.zeros(N-1)
    errors_v = np.zeros(N-1)
    for i in range(0,N-1):
        temp = resultsfile.readline().split()
        numbers[i] = i
        means_v[i] = temp[0]
        errors_v[i] = temp[1]

    plt.errorbar(numbers*throws, means_v, yerr=errors_v, errorevery=10)
    plt.xlabel('Total random throws')
    plt.ylabel('observed value of pi')
    plt.axhline(y=math.pi, color='red')
    plt.grid(True)
    plt.show()
```

Depending on the random generator seed used, the result is not always perfectly compatible with the expected value, but it is a good approximation, usually up to 2 significant digits.
